```python
%matplotlib notebook
import numpy as np
import crystaltools.crystal_generator as cg
import crystaltools.crystal_visualizer as cv
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import proj3d

#mpl.style.use('ggplot')
```

```python
unit = cg.get_unit_cell('fcc primitive', 2.8)
# unit = ct.get_unit_cell('bcc cubic', 2.8)
# unit = ct.get_unit_cell('fcc cubic', 2.8)
# sc = unit.create_supercell(10,10,10, name='supercell')
sc = unit.create_supercell(5, 5, 5, name='supercell')
# sc = unit.create_supercell(10,10,10, name='supercell')

v1 = np.array([1, 1, 1])
v2 = np.array([-1, 1, 1])

myplot = cv.CrystalPlot2D(lc='0.5', fheight=4., fwidth=8.)
myplot.ms = 80


fig = plt.figure(figsize=(4, 4))
fig.clf()
ax = fig.add_subplot(111)
myplot.plot_simple(sc, ax, v1, v2)
ax.axis('equal')

# v1 = np.array([1,0,0])
# v2 = np.array([0,1,0])
#myplot = cv.CrystalPlot3D(lc='0.5', ms=80)
#crys, fig, ax = myplot.plot_simple(sc, v1, v2, fignum=2);
```

```python
v1 = np.array([1,1,0])
v2 = np.array([0,1,0])
myplot = cv.CrystalPlot3D(lc='0.5', ms=40, fheight=8., fwidth=8.)

fig = plt.figure(figsize=(6, 8))
fig.clf()
ax = fig.add_subplot(111, projection='3d')

myplot.plot_simple(sc, ax, v1, v2, fignum=2);
```
